```python
from langchain import PromptTemplate

import hgutils
import time

stopwatch = hgutils.timer("Personal Shopper")
```

```python
categories = ['mobile','grocery']
details = ['reviews','product description',
            'comparison','delivery time',
            'alternate products','purchase',
            'recommendation']

products = ['iPhone 14 Pro','Samsung Galaxy S23 Ultra','Google Pixel 7']

moods = ['joking','formal','informal','fun','energetic']

def list_to_string(list,ordered=True,separator="\n"):
    s = ""
    for idx,list_item in enumerate(list):
        if ordered:
            s += f"{idx+1}.\t{list_item} \n"
        else:
            s += f"-.\t{list_item} \n"
    return s
```

```python
purchase_template = f'''
You are a personal shopper of a store named Tata Neu.
The store offers following products:
{list_to_string(products)}

The customer is seeking to purchase a product in below customer query, you job is to identify:
1. What product is the customer referring to? If customer is referring to more than one product mention all products separated by commas.
2. Your confidence on answer to question 1 in percentage
3. What is the mood of the customer?

Restrict your answer in following format:

{{{{product: answer to question 1,
product_confidence: answer to question 2,
mood: answer to question 3}}}}

Customer Query:
{{query}}
'''

purchase_prompt = PromptTemplate(input_variables=['query'],template=purchase_template)
query = "Hi I want to buy an iPhone"
prompt = purchase_prompt.format(query=query)
print(prompt)
```
